```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Covid Live.csv")
df.head()
```

```python
df.rename(columns={
    'Country,\nOther': 'Country',
    'Total\nCases': 'Cases',
    'Total\nDeaths': 'Deaths',
    'New\nDeaths' : 'New',
    'Total\nRecovered' : 'Totalrec'
    
}, inplace=True)
df
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python

ax = sns.countplot(data = df,x = "Country,\nOther")
ax.bar_label(ax.containers[0])
plt.title("country distrubution")
plt.show()
```

```python
df.sort_values(by='Total\nCases', ascending=False).head()
```

```python
df['CFR %'] = (df['Total\nDeaths'] / df['Total\nCases']) * 100
df
```

```python
df['Recovery %'] = (df['Total\nRecovered'] / df['Total\nCases']) * 100
df
```

```python
df.sort_values(by='Tests/\n1M pop', ascending=False).head(10)
df
```

```python
# Top 10 countries by total cases
top10 = df.sort_values(by='Cases', ascending=False).head(10)

plt.figure(figsize=(3,3))
sns.barplot(data=top10, x='Country', y='Cases')
plt.title('Top 10 Countries by Total Cases')
plt.xticks(rotation=45)
plt.show()
```
